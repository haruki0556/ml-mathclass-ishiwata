- 標準偏差（std）
データが平均値からどれだけ離れた位置まで分布しているか
偏差（平均との差）の絶対値の平均をとる（平均絶対偏差）MAD //一般には使わない
偏差（平均との差）の絶対値の二乗平均をとる（分散）
分散の平方根が標準偏差（std）

標準偏差は、多数派の範囲！！！　例、平均５０点のテストで標準偏差５なら、４５から５５点にたくさん人がいる
チェビシェフの不等式　mean +- k * std は、全体の1 - 1/k^2 以上 
標準偏差は外れ値に弱い（二乗するから、また平均で計算するから）

```python
data = np.array([1, 2, 3, 4, 5, 6])
print(f"平均値:{np.mean(data)},分散:{np.var(data)},標準偏差:{np.std(data)}")
print((1 - 1 / np.std(data) ** 2))
```

```python
x = np.array([-2, -1, 0, 1, 2])
y = np.array([4, 1, 0, 1, 4])
cov_matrix = np.cov(x, y)

print("共分散行列:")
print(cov_matrix)

cov_xy = np.cov(x, y)[0, 1]
print("共分散:", cov_xy)
```

```python
corr = np.corrcoef(x, y)[0, 1]
print("相関係数:", corr)
```
