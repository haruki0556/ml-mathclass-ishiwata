```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import MDS,TSNE 
```

```python
digits = load_digits()
X = digits.data
labels = digits.target
```

```python
X[0].reshape(8,8)
```

```python
labels[0]
```

```python
mds = MDS(n_components=2)
Ymds = mds.fit_transform(X)
Ymds
```

```python
plt.figure(figsize=(10,10))

for i in range(10):
    flag = labels==i
    Y_selected = Ymds[flag,:]

    plt.scatter(Y_selected[:,0],Y_selected[:,1],marker=f"${i}$")
plt.show()
```

# TSNE

```python
tsne = TSNE(n_components=2,perplexity = 50.0)
Y_tsne = tsne.fit_transform(X)

plt.figure(figsize=(10,10))
for i in range(10):
    flag = labels==i
    Y_selected = Y_tsne[flag,:]
    plt.scatter(Y_selected[:,0],Y_selected[:,1],marker=f"${i}$")

plt.show()
```
