```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.random.normal(loc=50, scale=10,size=10) # loc:h平均値　scale:標準偏差　size:数
x.mean()
```

```python
result = []
for i in range(10000):
    # x = np.random.normal(loc=50, scale=10, size=10)
    # x = np.random.exponential(scale=5, size=500)
    x = np.random.uniform(low=3, high=15, size=100) #　一様分布　指定範囲から同確率で　#平均9 分散12
    result.append(x.mean())

print("平均:",np.mean(result),"分散:",np.var(result)) #中心極限定理から、平均は9付近、分散は12/sizeに収束
print(np.array(result)>50+2*np.sqrt(10))
plt.hist(result)
plt.show()

#　標本分布の分散(母分散/size)が小さくなっていれば、sizeが十分と判断できる
```

```python
print(np.mean(np.array(result)>50+2*np.sqrt(10))) #50 
```

```python
result = []
for i in range(10000):

    x = np.random.normal(loc=50, scale=10, size=10) #母分散100 
    result.append(x.var())

print("平均:",np.mean(result)) #分散の平均

plt.hist(result)
plt.show()

#　標本分散と母分散が一致せず、、、
```

```python
result1 = []
for i in range(10000):

    x = np.random.normal(loc=50, scale=10, size=10) #母分散100 
    result1.append(x.var(ddof=1)) #ddof 10-1(引数)で割る
print("平均:",np.mean(result1)) 

plt.hist(result1)
plt.show()
"""
ddof で1少なく割るとほぼほぼ一致(不遍分散)
[標本分散の標本分布の平均が、母平均と一致]という文脈で不偏性
"""
```
