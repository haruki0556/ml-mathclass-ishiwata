```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import DBSCAN#スケーリングも一般には推奨
from sklearn.datasets import make_moons
```

```python
X,y = make_moons(100,noise=0.05)

plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
model = DBSCAN(eps=0.2,min_samples=3)

cluster = model.fit_predict(X)
cluster
```

```python
for i in np.unique(cluster):
    index = cluster==i

    plt.scatter(X[index,0],X[index,1],label = i)

plt.legend()
plt.show()
```

```python
def k_distance(x,k):
    dim0 = x.shape[0]#サンプルサイズをとる

    p = -2*x@x.T + np.sum(x**2,axis=1).T + np.repeat(np.sum(x**2,axis=1),dim0,axis=0).reshape(dim0,dim0)#距離行列の作成
    p = np.sqrt(p)
    p.sort(axis=1)
    p = p[:,k]
    p.sort()

    return p

```
